```python
from dataclasses import dataclass

@dataclass
class Packet:
    source_ip: str
    dest_ip: str
    payload: str
    priority: int
```

```python
def fifo_scheduler(packet_list: list) -> list:
    return packet_list.copy()
```

```python
def priority_scheduler(packet_list: list) -> list:
    sorted_packets = sorted(packet_list, key=lambda p: p.priority)
    return sorted_packets
```

```python
packets = [
    Packet("192.168.1.1", "10.0.0.1", "Data Packet 1", 2),
    Packet("192.168.1.2", "10.0.0.2", "Data Packet 2", 2),
    Packet("192.168.1.3", "10.0.0.3", "VOIP Packet 1", 0),
    Packet("192.168.1.4", "10.0.0.4", "Video Packet 1", 1),
    Packet("192.168.1.5", "10.0.0.5", "VOIP Packet 2", 0)
]

print("Testing FIFO Scheduler:")
fifo_result = fifo_scheduler(packets)
print([p.payload for p in fifo_result])
print()

print("Testing Priority Scheduler:")
priority_result = priority_scheduler(packets)
print([p.payload for p in priority_result])
```
