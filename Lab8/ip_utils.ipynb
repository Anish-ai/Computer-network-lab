```python
def ip_to_binary(ip_address: str) -> str:
    octets = ip_address.split('.')
    binary_octets = []
    
    for octet in octets:
        binary = bin(int(octet))[2:]
        binary_octets.append(binary.zfill(8))
    
    return ''.join(binary_octets)
```

```python
def get_network_prefix(ip_cidr: str) -> str:
    ip_part, prefix_length = ip_cidr.split('/')
    prefix_length = int(prefix_length)
    
    binary_ip = ip_to_binary(ip_part)
    network_prefix = binary_ip[:prefix_length]
    
    return network_prefix
```

```python
print("Testing ip_to_binary:")
print(f"192.168.1.1 -> {ip_to_binary('192.168.1.1')}")
print(f"223.1.1.100 -> {ip_to_binary('223.1.1.100')}")
print()

print("Testing get_network_prefix:")
print(f"200.23.16.0/23 -> {get_network_prefix('200.23.16.0/23')}")
print(f"223.1.1.0/24 -> {get_network_prefix('223.1.1.0/24')}")
print(f"223.1.0.0/16 -> {get_network_prefix('223.1.0.0/16')}")
```
